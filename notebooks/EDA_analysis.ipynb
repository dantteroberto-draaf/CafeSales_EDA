```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_cafe = pd.read_csv(r'C:\Users\dantt\Documents\Códigos\Data Science e ML - Udemy\CafeSales_EDA\data\processed\clean_cafe_sales.csv')
```

## Insights

Agora, com os dados devidamente tratados, podemos criar gráficos e obter insights dessas informações

#### **Vendas por item**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Item', data=base_cafe, palette='viridis', hue='Item',  order=base_cafe['Item'].value_counts().index)

# Estética 
plt.title('Contagem de Vendas por Item', fontsize=14)
plt.xlabel('Produto')
plt.ylabel('Número de Transações')
plt.xticks(rotation=45) 
plt.show()
```

É possível visualizar, a partir do gráfico de barras, que o item mais vendido é o Coffee, fato que tem relação direta com o tipo de estabelecimento(cafeteria). Porém, logo após o café, vem a Salad e o Cookie. Sob esse ponto de vista, nota-se que, frequentemente, os clientes não querem apenas tomar café mas também fazer alguma refeição ou comprar um lanche junto com a bebida. Essa informação é valiosa, pois permite ao estabelecimento montar estratégias para vender, em conjunto, mais de um item e, dessa forma, aumentar sua amplitude de vendas.

---

#### Gráfico Boxplot **(Gasto total x Método de pagamento)**

```python
plt.figure(figsize=(10, 6))

# x = Categoria (eixo horizontal), y = Número (eixo vertical)
sns.boxplot(x='Payment Method', y='Total Spent', data=base_cafe, palette='coolwarm', hue='Payment Method')

plt.title('Distribuição de Gastos por Método de Pagamento', fontsize=14)
plt.show()
```

Analisando o Boxplot, é possível perceber que, independente do método de pagamento, o ticket médio é o mesmo. Ou seja, os meios de pagamento são igualmente utilizados e não há uma preferência dos clientes sobre qual método usar. O valor total da compra não influencia na escolha de meio de pagamento da pessoa que realiza a compra.

---

#### Relação **Item x Preferência de consumo**

```python
plt.figure(figsize=(12, 6))

sns.countplot(x='Item', hue='Location', data=base_cafe[base_cafe['Location'] != 'UNKNOWN'], palette='pastel')

plt.title('Preferência de Consumo por Item (Loja vs. Viagem)', fontsize=14)
plt.legend(title='Local de Consumo')
plt.xticks(rotation=45)
plt.show()
```

A partir deste gráfico, é possível perceber que os produtos com maior discrepância entre o consumo em loja e o takeaway são o Café o Cookie. Em sua maioria, esses produtos são comprados para serem consumidos fora do estabelecimento; isso indica que os produtos de tamanho menor são compras feitas rapidamente pelos clientes, que pegam o produto e saem da cafeteria em pouco tempo. Já os produtos como Salada e Sanduíche retém os clientes por mais tempo na cafeteria, pois são majoritariamente consumidos dentro do estabelecimento.

Nesse prisma, a cafeteria poderia elaborar combos que unissem os itens de acordo com a preferência de consumo, visando aumentar o faturamento. Por exemplo: um combo Salad + Juice, já que são frequentemente consumidos dentro da loja; um combo Cookie + Coffee, já que são frequentemente consumidos fora da loja.
